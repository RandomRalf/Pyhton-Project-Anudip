```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load your data into a DataFrame
df = pd.read_excel('Online Retail Data Set.xlsx')
print(df)
```

```python
# Perform Data Cleaning

# Removing null values from the data
df=df.dropna()
print(df)
```

```python
#Looking at the new cleaned data
df.info()
```

```python
# Divide the data in half to make it smaller

# Get the number of rows in the DataFrame
num_rows = len(df)

# Calculate the midpoint
midpoint = num_rows // 2

# Split the DataFrame into two halves
df = df.iloc[:midpoint]

#Looking at the new cleaned data
df.info()
```

```python
# Drop useless column
df = df.drop(columns=['Description'])
df = df.drop(columns=['InvoiceNo'])
# Looking at the new cleaned data
df.info()
```

```python
# Create a graph to show number of customer by each country
plt.figure(figsize=(20, 7))
plt.bar(df['Country'], df['CustomerID'])
plt.xlabel('Country')
plt.ylabel('number of Customer')
plt.title('Countries by number of Customer')
plt.show()
```

```python
# Create a graph to show top 5 number of customer by each country

# Group by country and find the maximum CustomerID for each country
max_customer_id_by_country = df.groupby('Country')['CustomerID'].max()

# Select the top 5 countries with the highest CustomerID
top_5_countries = max_customer_id_by_country.nlargest(5)

# Create a bar graph
plt.figure(figsize=(10, 7))
plt.bar(top_5_countries.index, top_5_countries.values)
plt.xlabel('Country')
plt.ylabel('number of Customer')
plt.title('Top 5 Countries by number of Customer')
plt.show()
```

```python
# Create a graph to find out top 5 countries with the most income

# Group by country and find the maximum UnitPrice for each country
max_unit_price_by_country = df.groupby('Country')['UnitPrice'].max()

# Select the top 5 countries with the highest UnitPrice
top_5_countries = max_unit_price_by_country.nlargest(5)

# Create a bar graph
plt.figure(figsize=(7, 7))
plt.bar(top_5_countries.index, top_5_countries.values)
plt.xlabel('Country')
plt.ylabel('income')
plt.title('Top 5 Countries by income')
plt.show()

# Create a pie chart 
plt.figure(figsize=(10, 7)) 
top_5_countries.plot.pie(autopct='%1.1f%%', startangle=90) 
plt.title('Top 5 countries Income Distribution') 
plt.ylabel('') 
plt.show()
```

```python
# Create a graph with sales for each month

# Convert the InvoiceDate column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create a new column for the month and year
df['MonthYear'] = df['InvoiceDate'].dt.to_period('M')

# Calculate the total sales for each row
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Group by the new MonthYear column and sum the TotalSales
monthly_sales = df.groupby('MonthYear')['TotalSales'].sum()

# Create a pie chart 
plt.figure(figsize=(10, 7)) 
monthly_sales.plot.pie(autopct='%1.1f%%', startangle=90) 
plt.title('Monthly Sales Distribution') 
plt.ylabel('') # Hide the y-label
```

```python
# Create a graph with sales from each month on the country with most sales 

# Convert the InvoiceDate column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Filter the data for only the United Kingdom
df_uk = df[df['Country'] == 'United Kingdom']

# Group by the new MonthYear column and sum the TotalSales
monthly_sales_uk = df_uk.groupby('MonthYear')['TotalSales'].sum()

# Create a pie chart 
plt.figure(figsize=(10, 7))
monthly_sales_uk.plot.pie(autopct='%1.1f%%', startangle=90) 
plt.title('Monthly Sales Distribution for United Kingdom') 
plt.ylabel('')
plt.show()
```

```python
# Create a Graph with top 10 customer with the most transaction

# Group by CustomerID and find the maximum UnitPrice for each customer
max_unitprice_by_customer = df.groupby('CustomerID')['UnitPrice'].max().reset_index()

# Select the top 10 customers with the highest UnitPrice
top_10_customers = max_unitprice_by_customer.nlargest(10, 'UnitPrice')

# Create a bar graph
plt.figure(figsize=(10, 7))
plt.bar(top_10_customers['CustomerID'].astype(str), top_10_customers['UnitPrice'])
plt.xlabel('CustomerID')
plt.ylabel('UnitPrice')
plt.title('Top 10 Customers with the most income generated')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a graph with top 10 customer with the number of sale.

# Group by CustomerID and count the number of unique InvoiceNo for each customer.
invoice_counts_by_customer = df.groupby('CustomerID')['StockCode'].nunique().reset_index()

# Select the top 10 customers with the highest number of InvoiceNo
top_10_customers = invoice_counts_by_customer.nlargest(10, 'StockCode')

# Create a bar graph
plt.figure(figsize=(10, 7))
plt.bar(top_10_customers['CustomerID'].astype(str), top_10_customers['StockCode'])
plt.xlabel('CustomerID')
plt.ylabel('Number of Invoices')
plt.title('Top 10 Customers with the most number of sale')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a graph for top customer for each month

# Calculate the total sales for each row
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Group by MonthYear and CustomerID and sum the TotalSales
monthly_sales = df.groupby(['MonthYear', 'CustomerID'])['TotalSales'].sum().reset_index()

# Find the top income-generating customer for each month
top_customers = monthly_sales.loc[monthly_sales.groupby('MonthYear')['TotalSales'].idxmax()]
# Print top customers
print(top_customers)
# Create a bar graph
plt.figure(figsize=(12, 8))
plt.bar(top_customers['MonthYear'].astype(str), top_customers['TotalSales'])
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.title('Top Income Generating Customer Each Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a graph for most sold item of all time

# Group by StockCode and sum the Quantity for each stock item
total_quantity_by_stock = df.groupby('StockCode')['Quantity'].sum().reset_index()

# Select the top 5 most sold stock items
top_5_stock_items = total_quantity_by_stock.nlargest(5, 'Quantity')

# Convert StockCode to string 
top_5_stock_items['StockCode'] = top_5_stock_items['StockCode'].astype(str)

# Create a bar graph
plt.figure(figsize=(10, 7))
plt.bar(top_5_stock_items['StockCode'], top_5_stock_items['Quantity'])
plt.xlabel('StockCode')
plt.ylabel('Total Quantity Sold')
plt.title('Top 5 Most Sold Stock Items')
plt.show()
```
